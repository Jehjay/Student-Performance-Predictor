CREATE 
1. A FUNCTION TO CHANGE THE [G1, G2] columns ACCORDING TO A RANGE based on the Erasmus grade conversion system
2. A FUNCTION FOR ONE HOT ENCONDING THE ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic'] columns.

```python
def student_performance_preprocess_onehot(X):
    X_math = X 
    #convert to bool
    to_bool = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
    # function to convert to bool
    def convert_fill_bool (val):
        if val == 'yes': return 1
        elif val == 'Yes': return 1
        else: return 0
        
    for col in to_bool:
        X_math[col] = X_math[col].apply(convert_fill_bool)
    return X_math
```

```python
X_math = student_performance_preprocess_onehot(X_math)
Stundent_math_course_df = pd.concat([y_math,X_math], axis = 'columns')
# .dropna()
Student_math_course_df.head()
```

```python
def student_performance_preprocess_grade(X):
    X_math = X 
    # grading columns
    grade = ['G1', 'G2']
    def erasmus_grade_range (val):
        if (0<val<=9) : return 5
        if (10<val<=11) : return 4
        if (12<val<=13) : return 3
        if (14<val<=15) : return 2
        if (15<val<=20) : return 1
        else: return np.nan
        
    for column in grade:
        X_math[column] = X_math[column].apply(erasmus_grade_range)
    return X_math
```

```python
X_math = student_performance_preprocess_grade(X_math)
Student_math_course_df = pd.concat([y_math,X_math], axis = 'columns')
# .dropna()
Student_math_course_df.head()
```

```python
len(predicted_success_math)
```
